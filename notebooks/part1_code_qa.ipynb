```python
from src.part1.pipeline import classify_query, TEST_QUESTIONS

print(f'{"#":<3} {"Label":<18} Question')
print('-' * 85)
for i, q in enumerate(TEST_QUESTIONS, 1):
    label = classify_query(q)
    print(f'Q{i:<2} {str(label):<18} {q[:62]}')
```

```python
from src.part1.pipeline import answer_question

def run_q(n: int) -> 'QAResult':
    q = TEST_QUESTIONS[n - 1]
    print(f'\n{"═" * 72}')
    print(f'Q{n}: {q}')
    print('─' * 72)
    result = answer_question(REPO_ROOT, q)
    print(f'label    : {classify_query(q)}')
    print()
    print(result.answer)
    return result
```
